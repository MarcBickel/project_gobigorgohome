```python
number_of_col = []
for i in range(175):
    number_of_col.append(i)
```

```python
in_path = "data/en.openfoodfacts.org.products.tar.xz" #Path where the large file is
out_path = "data" #Path to save the pickle files to
chunksize = 100000 #size of chunks relies on your available memory

i=0
for chunk in pd.read_csv(in_path, sep='\t', usecols=number_of_col, chunksize=chunksize):
    out_file = out_path + "/off_chunk_{}.pkl".format(i+1)
    i+=1
    with open(out_file, "wb") as f:
        pickle.dump(chunk,f,pickle.HIGHEST_PROTOCOL)

    
```

```python
import glob
pickle_path = "data" #Same Path as out_path i.e. where the pickle files are

data_p_files=[]
for name in glob.glob(pickle_path + "/off_chunk_*.pkl"):
   data_p_files.append(name)


df = pd.DataFrame([])
for i in range(len(data_p_files)):
    df = df.append(pd.read_pickle(data_p_files[i]),ignore_index=True)
```
